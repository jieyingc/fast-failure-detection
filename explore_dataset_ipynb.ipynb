```python
from datasets import load_dataset

dataset = load_dataset("SWE-bench/SWE-smith-trajectories")

print(dataset)
```

```python
def check_nulls(hf_dataset):
    for col in hf_dataset.column_names:
        null_count = sum(row[col] is None for row in hf_dataset)
        print(f"{col}: {null_count} nulls")

check_nulls(dataset["tool"])
```
