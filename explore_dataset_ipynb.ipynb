```python
def check_empty_strings(hf_dataset):
    for col in hf_dataset.column_names:
        empty_count = 0
        for row in hf_dataset:
            value = row[col]
            if isinstance(value, str) and value.strip() == "":
                empty_count += 1
        print(f"{col}: {empty_count} empty strings")

print(check_empty_strings(dataset["tool"]))
print(check_empty_strings(dataset["xml"]))
print(check_empty_strings(dataset["ticks"]))
```

**Feature extraction**

```python
import json
import re


def extract_features(row):
    patch = row["patch"] if row["patch"] is not None else ""
    patch_stripped = patch.strip()

    try:
        msgs = json.loads(row["messages"])
    except Exception:
        msgs = []

    # -------------------------
    # 1) Patch-level features
    # -------------------------
    patch_is_empty = int(patch_stripped == "")
    patch_char_len = len(patch)

    # Count added / removed lines (excluding diff headers like +++ / ---)
    num_lines_added = 0
    num_lines_removed = 0

    for line in patch.splitlines():
        if line.startswith("+++ ") or line.startswith("--- "):
            continue
        if line.startswith("+"):
            num_lines_added += 1
        elif line.startswith("-"):
            num_lines_removed += 1

    # Number of files touched
    num_files_touched = patch.count("diff --git")

    # Whether test files were modified
    touched_files = re.findall(r"diff --git a/(.*?) b/(.*?)\n", patch)
    test_file_modified = 0
    for a_path, b_path in touched_files:
        path = b_path.lower()
        if (
            "test" in path
            or "tests/" in path
            or path.endswith("_test.py")
            or "/test_" in path
        ):
            test_file_modified = 1
            break

    # Repeated-block (if the same added/removed code line repeats 3 times)
    code_lines = []
    for line in patch.splitlines():
        if line.startswith("+++ ") or line.startswith("--- ") or line.startswith("@@"):
            continue
        if line.startswith("+") or line.startswith("-"):
            content = line[1:].strip()
            if content:
                code_lines.append(content)

    contain_repeated_blocks = 0
    if code_lines:
        line_freq = {}
        for x in code_lines:
            line_freq[x] = line_freq.get(x, 0) + 1
        if max(line_freq.values()) >= 3:
            contain_repeated_blocks = 1

    # -------------------------
    # 2) Message / trajectory features
    # -------------------------
    num_messages = len(msgs)
    messages_char_len = len(row["messages"])

    num_assistant_messages = 0
    num_user_messages = 0
    num_tool_messages = 0
    num_system_messages = 0

    num_action_messages = 0
    num_observation_messages = 0

    has_submit = 0
    num_submit_calls = 0

    has_bash = 0
    num_bash_calls = 0

    has_str_replace_editor = 0

    num_edit_calls = 0
    num_error_messages = 0

    edit_tool_names = {"str_replace_editor", "edit_file", "create_file", "insert", "replace", "write_file"}

    error_keywords = ["error", "exception", "traceback", "failed", "failure", "syntaxerror", "typeerror", "valueerror", "assertionerror"]

    for msg in msgs:
        # Role counts
        role = msg.get("role", "")

        if role == "assistant":
            num_assistant_messages += 1
        elif role == "user":
            num_user_messages += 1
        elif role == "tool":
            num_tool_messages += 1
        elif role == "system":
            num_system_messages += 1

        # Message-type counts
        message_type = msg.get("message_type", "")

        if message_type == "action":
            num_action_messages += 1
        elif message_type == "observation":
            num_observation_messages += 1

        # Check action field
        action_text = str(msg.get("action", "")).lower()

        if "submit" in action_text:
            has_submit = 1

        if "bash" in action_text:
            has_bash = 1

        if "str_replace_editor" in action_text:
            has_str_replace_editor = 1

        # tool_calls parsing

        tool_calls = msg.get("tool_calls", [])
        if isinstance(tool_calls, list):
            for tc in tool_calls:
                fn_name = (
                    tc.get("function", {}).get("name", "")
                    if isinstance(tc, dict)
                    else ""
                )
                fn_name = str(fn_name).lower()

                if fn_name == "submit":
                    has_submit = 1
                    num_submit_calls += 1

                if fn_name == "bash":
                    has_bash = 1
                    num_bash_calls += 1

                if fn_name == "str_replace_editor":
                    has_str_replace_editor = 1

                if fn_name in edit_tool_names or fn_name == "str_replace_editor":
                    num_edit_calls += 1

        # error keyword counting

        text_blobs = [] # Collect all text (content, thought, action)

        content = msg.get("content", "")
        if isinstance(content, str):
            text_blobs.append(content.lower())
        elif isinstance(content, list):
            for item in content:
                if isinstance(item, dict):
                    t = item.get("text", "")
                    if isinstance(t, str):
                        text_blobs.append(t.lower())

        thought = msg.get("thought", "")
        if isinstance(thought, str):
            text_blobs.append(thought.lower())

        if isinstance(action_text, str):
            text_blobs.append(action_text)

        joined_text = " ".join(text_blobs)
        for kw in error_keywords:
            num_error_messages += joined_text.count(kw)


    features = {
        # IDs / metadata
        "instance_id": row["instance_id"],
        "traj_id": row["traj_id"],
        "model": row["model"],

        # Label
        "resolved": int(bool(row["resolved"])),

        # Patch features
        "patch_is_empty": patch_is_empty,
        "patch_char_len": patch_char_len,
        "num_lines_added": num_lines_added,
        "num_lines_removed": num_lines_removed,
        "num_files_touched": num_files_touched,
        "test_file_modified": test_file_modified,
        "contain_repeated_blocks": contain_repeated_blocks,

        # Message features
        "num_messages": num_messages,
        "messages_char_len": messages_char_len,
        "assistant_num_messages": num_assistant_messages,
        "user_num_messages": num_user_messages,
        "tool_num_messages": num_tool_messages,
        "system_num_messages": num_system_messages,
        "num_action_messages": num_action_messages,
        "num_observation_messages": num_observation_messages,
        "has_submit": has_submit,
        "num_submit_calls": num_submit_calls,
        "has_bash": has_bash,
        "num_bash_calls": num_bash_calls,
        "has_str_replace_editor": has_str_replace_editor,
        "num_edit_calls": num_edit_calls,
        "num_error_messages": num_error_messages,
    }

    return features
```

```python
import pandas as pd
from tqdm import tqdm

feature_rows = []

for i in tqdm(range(len(dataset["tool"]))):
    row = dataset["tool"][i]
    feature_rows.append(extract_features(row))

df = pd.DataFrame(feature_rows)

print(df.shape)
df.head()
```

```python
print(df["resolved"].value_counts())
print(df["patch_is_empty"].value_counts())
print(df.isnull().sum().sort_values(ascending=False).head(10))
```

```python
from sklearn.model_selection import GroupShuffleSplit

# train (80%) / val (10%) / test (10%)
gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
train_idx, temp_idx = next(gss.split(df, groups=df["instance_id"]))

train_df = df.iloc[train_idx].reset_index(drop=True)
temp_df = df.iloc[temp_idx].reset_index(drop=True)


gss2 = GroupShuffleSplit(n_splits=1, test_size=0.5, random_state=42)
val_idx, test_idx = next(gss2.split(temp_df, groups=temp_df["instance_id"]))

val_df = temp_df.iloc[val_idx].reset_index(drop=True)
test_df = temp_df.iloc[test_idx].reset_index(drop=True)

print("train shape:", train_df.shape)
print("val shape:", val_df.shape)
print("test shape:", test_df.shape)
```

```python
train_instances = set(train_df["instance_id"])
val_instances = set(val_df["instance_id"])
test_instances = set(test_df["instance_id"])

print("train ∩ val:", len(train_instances & val_instances))
print("train ∩ test:", len(train_instances & test_instances))
print("val ∩ test:", len(val_instances & test_instances))
```

```python
import numpy as np

log_features = [
    "patch_char_len",
    "num_lines_added",
    "num_lines_removed",
    "num_files_touched",
    "num_messages",
    "messages_char_len",
    "num_submit_calls",
    "num_bash_calls",
    "num_edit_calls",
    "num_error_messages",
]

for col in log_features:
    train_df[col] = np.log1p(train_df[col])
    val_df[col] = np.log1p(val_df[col])
    test_df[col] = np.log1p(test_df[col])
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

feature_cols = [
    "patch_is_empty",
    "patch_char_len",
    "num_lines_added",
    "num_lines_removed",
    "num_files_touched",
    "test_file_modified",
    "contain_repeated_blocks",
    "num_messages",
    "messages_char_len",
    "assistant_num_messages",
    "user_num_messages",
    "tool_num_messages",
    "system_num_messages",
    "num_action_messages",
    "num_observation_messages",
    "has_submit",
    "num_submit_calls",
    "has_bash",
    "num_bash_calls",
    "has_str_replace_editor",
    "num_edit_calls",
    "num_error_messages",
]

X_train = train_df[feature_cols]
y_train = train_df["resolved"]

X_val = val_df[feature_cols]
y_val = val_df["resolved"]

X_test = test_df[feature_cols]
y_test = test_df["resolved"]

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

clf = LogisticRegression(
    max_iter=2000,
    class_weight="balanced",
    random_state=42
)

clf.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, classification_report

val_pred = clf.predict(X_val_scaled)
val_prob = clf.predict_proba(X_val_scaled)[:, 1]

print("Validation Accuracy:", accuracy_score(y_val, val_pred))
print("Validation F1:", f1_score(y_val, val_pred))
print("Validation ROC-AUC:", roc_auc_score(y_val, val_prob))
print(classification_report(y_val, val_pred))
```

```python
test_pred = clf.predict(X_test_scaled)
test_prob = clf.predict_proba(X_test_scaled)[:, 1]

print("Test Accuracy:", accuracy_score(y_test, test_pred))
print("Test F1:", f1_score(y_test, test_pred))
print("Test ROC-AUC:", roc_auc_score(y_test, test_prob))
print(classification_report(y_test, test_pred))
```
