```python
from collections import Counter

print(Counter(dataset["tool"]["resolved"]))
print(Counter(dataset["xml"]["resolved"]))
print(Counter(dataset["ticks"]["resolved"]))
```

```python
tool_ids = set(dataset["tool"]["instance_id"])
xml_ids = set(dataset["xml"]["instance_id"])
ticks_ids = set(dataset["ticks"]["instance_id"])

print("tool ∩ xml:", len(tool_ids & xml_ids))
print("tool ∩ ticks:", len(tool_ids & ticks_ids))
print("xml ∩ ticks:", len(xml_ids & ticks_ids))
```

### **Check null and empty**

```python
for col in dataset["tool"].column_names:
    null_count = sum(row[col] is None for row in dataset["tool"])
    print(f"{col}: {null_count} nulls")
```

```python
def check_empty_strings(hf_dataset):
    for col in hf_dataset.column_names:
        empty_count = 0
        for row in hf_dataset:
            value = row[col]
            if isinstance(value, str) and value.strip() == "":
                empty_count += 1
        print(f"{col}: {empty_count} empty strings")

print(check_empty_strings(dataset["tool"]))
print(check_empty_strings(dataset["xml"]))
print(check_empty_strings(dataset["ticks"]))
```

```python
empty_patch_labels = Counter()
nonempty_patch_labels = Counter()

for row in dataset["tool"]:
    if row["patch"].strip() == "":
        empty_patch_labels[row["resolved"]] += 1
    else:
        nonempty_patch_labels[row["resolved"]] += 1

print("Empty patch labels:", empty_patch_labels)
print("Non-empty patch labels:", nonempty_patch_labels)
```

**split by instance_id not by row**:

```python
tool_instance_counts = Counter(dataset["tool"]["instance_id"])
dup_tool = sum(c > 1 for c in tool_instance_counts.values())
max_count = max(tool_instance_counts.values())

print("unique instance_ids:", len(tool_instance_counts))
print("duplicated instance_ids:", dup_tool)
print("max trajectories per instance:", max_count)
```
