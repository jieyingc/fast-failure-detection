```python
import sys
import json
import re
from pathlib import Path

import numpy as np
import pandas as pd
from datasets import load_dataset
```

```python
PROJECT_ROOT = Path("/content/fast-failure-detection")
RESULTS_DIR = Path("/content/drive/MyDrive/fast-failure-detection/results")
RESULTS_DIR.mkdir(parents=True, exist_ok=True)

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

print(PROJECT_ROOT)
print(RESULTS_DIR)
```

```python
traj_df = load_dataset("SWE-bench/SWE-smith-trajectories", split="tool").to_pandas()
smith_df = load_dataset("SWE-bench/SWE-smith-py", split="train").to_pandas()

print("traj_df:", traj_df.shape)
print("smith_df:", smith_df.shape)
```

```python
def is_empty_patch(x):
    if pd.isna(x):
        return True
    s = str(x).strip()
    return s == "" or s.lower() in {"nan", "none", "null"}

traj_df["patch_is_empty"] = traj_df["patch"].apply(is_empty_patch)
traj_df = traj_df.loc[~traj_df["patch_is_empty"]].copy()

print("filtered traj_df:", traj_df.shape)
```

```python
merged_df = traj_df.merge(
    smith_df,
    on="instance_id",
    how="left",
    suffixes=("_traj", "_smith")
)

analysis_df = merged_df.loc[merged_df["repo"].notna()].copy()
analysis_df["repo_id"] = analysis_df["instance_id"].astype(str).str.split(".", n=1).str[0]

print("analysis_df:", analysis_df.shape)
print("missing repo:", analysis_df["repo"].isna().mean())
```

```python
def to_py_list(x):
    if x is None:
        return []
    if isinstance(x, np.ndarray):
        return x.tolist()
    if isinstance(x, (list, tuple)):
        return list(x)
    if isinstance(x, float) and np.isnan(x):
        return []
    return []

analysis_df["ftp_list"] = analysis_df["FAIL_TO_PASS"].apply(to_py_list)
analysis_df["ptp_list"] = analysis_df["PASS_TO_PASS"].apply(to_py_list)
```

```python
def parse_test_id(test_id: str):
    """
    Input like:
    tests/test_core.py::TestPath::test_xxx[param]
    """
    if not isinstance(test_id, str):
        return {"file": None, "class": None, "test": None}

    s = test_id.strip()
    if not s:
        return {"file": None, "class": None, "test": None}

    parts = s.split("::")
    file_part = parts[0] if len(parts) >= 1 else None
    class_part = parts[1] if len(parts) >= 3 else None
    test_part = parts[-1] if len(parts) >= 2 else None

    # 去掉参数化部分 [xxx]
    if isinstance(test_part, str):
        test_part = re.sub(r"\[.*\]$", "", test_part)

    return {
        "file": file_part,
        "class": class_part,
        "test": test_part,
    }

def collect_gold_sets(test_list):
    files, classes, tests = set(), set(), set()
    for x in test_list:
        parsed = parse_test_id(x)
        if parsed["file"]:
            files.add(parsed["file"].lower())
        if parsed["class"]:
            classes.add(parsed["class"])
        if parsed["test"]:
            tests.add(parsed["test"])
    return {
        "files": files,
        "classes": classes,
        "tests": tests,
    }
```

```python
EDIT_TOOL_NAMES = {
    "str_replace_editor",
    "edit_file",
    "create_file",
    "insert",
    "replace",
    "write_file",
}

def normalize_text(s: str):
    s = s.lower().strip()
    s = re.sub(r"\s+", " ", s)
    return s

def extract_path_like_strings(text: str):
    """
    粗略抓路径:
    - xxx/yyy.py
    - tests/test_x.py
    """
    if not isinstance(text, str):
        return set()
    candidates = re.findall(r'[\w\-/\.]+(?:\.py|\.js|\.ts|\.java|\.go|\.rs|\.cpp|\.php)', text)
    return {c.lower() for c in candidates if "/" in c or c.endswith(".py")}

def extract_edit_path_from_tool_call(tc):
    if not isinstance(tc, dict):
        return None
    fn = tc.get("function", {})
    args = fn.get("arguments", "")
    if not isinstance(args, str):
        return None

    patterns = [
        r'"path"\s*:\s*"([^"]+)"',
        r'"file_path"\s*:\s*"([^"]+)"',
        r'"filename"\s*:\s*"([^"]+)"',
        r'"target_file"\s*:\s*"([^"]+)"',
    ]
    for pat in patterns:
        m = re.search(pat, args)
        if m:
            return m.group(1).strip().lower()
    return None

def flatten_message_text(msg):
    content = msg.get("content", "")
    texts = []

    if isinstance(content, str):
        texts.append(content)
    elif isinstance(content, list):
        for item in content:
            if isinstance(item, dict):
                t = item.get("text", "")
                if isinstance(t, str):
                    texts.append(t)

    return "\n".join(texts)

def parse_messages(messages_json):
    """
    返回:
    - all_text
    - all_paths_from_text
    - edit_paths
    - bash_texts
    - step_records: 每步的 text/paths
    """
    try:
        msgs = json.loads(messages_json)
    except Exception:
        msgs = []

    all_text_parts = []
    all_paths_from_text = set()
    edit_paths = set()
    bash_texts = []
    step_records = []

    for step_idx, msg in enumerate(msgs):
        text = flatten_message_text(msg)
        text_norm = normalize_text(text) if isinstance(text, str) else ""
        text_paths = extract_path_like_strings(text_norm)

        all_text_parts.append(text_norm)
        all_paths_from_text |= text_paths

        tc_paths = set()
        tc_bash_texts = []

        tool_calls = msg.get("tool_calls", [])
        if isinstance(tool_calls, list):
            for tc in tool_calls:
                if not isinstance(tc, dict):
                    continue
                fn = tc.get("function", {})
                fn_name = str(fn.get("name", "")).lower()
                fn_args = str(fn.get("arguments", ""))

                if fn_name in EDIT_TOOL_NAMES:
                    p = extract_edit_path_from_tool_call(tc)
                    if p:
                        edit_paths.add(p)
                        tc_paths.add(p)

                if fn_name == "bash":
                    tc_bash_texts.append(fn_args)
                    bash_texts.append(fn_args)
                    all_paths_from_text |= extract_path_like_strings(fn_args.lower())

        step_records.append({
            "step_idx": step_idx,
            "text": text_norm,
            "text_paths": text_paths,
            "tool_paths": tc_paths,
            "bash_texts": tc_bash_texts,
        })

    all_text = " ".join(all_text_parts)

    return {
        "all_text": all_text,
        "all_paths_from_text": all_paths_from_text,
        "edit_paths": edit_paths,
        "bash_texts": bash_texts,
        "step_records": step_records,
    }
```

```python
def overlap_ratio(hit_set, gold_set):
    if len(gold_set) == 0:
        return np.nan
    return len(hit_set & gold_set) / len(gold_set)

def first_hit_step(step_records, matcher_fn):
    for rec in step_records:
        if matcher_fn(rec):
            return rec["step_idx"]
    return -1

def make_alignment_features(row):
    ftp_sets = collect_gold_sets(row["ftp_list"])
    ptp_sets = collect_gold_sets(row["ptp_list"])

    msg_info = parse_messages(row["messages"])

    all_text = msg_info["all_text"]
    all_paths = set(msg_info["all_paths_from_text"]) | set(msg_info["edit_paths"])
    step_records = msg_info["step_records"]

    # -------------------------
    # Level 1: 文件级
    # -------------------------
    ftp_file_hits = {p for p in ftp_sets["files"] if p in all_paths}
    ptp_file_hits = {p for p in ptp_sets["files"] if p in all_paths}

    has_ftp_file_overlap = int(len(ftp_file_hits) > 0)
    has_ptp_file_overlap = int(len(ptp_file_hits) > 0)

    ftp_file_overlap_count = len(ftp_file_hits)
    ptp_file_overlap_count = len(ptp_file_hits)

    ftp_file_overlap_ratio = overlap_ratio(ftp_file_hits, ftp_sets["files"])
    ptp_file_overlap_ratio = overlap_ratio(ptp_file_hits, ptp_sets["files"])

    # -------------------------
    # Level 2: class/test name级
    # -------------------------
    ftp_class_hits = {c for c in ftp_sets["classes"] if c.lower() in all_text}
    ptp_class_hits = {c for c in ptp_sets["classes"] if c.lower() in all_text}

    ftp_test_hits = {t for t in ftp_sets["tests"] if t.lower() in all_text}
    ptp_test_hits = {t for t in ptp_sets["tests"] if t.lower() in all_text}

    has_ftp_class_overlap = int(len(ftp_class_hits) > 0)
    has_ptp_class_overlap = int(len(ptp_class_hits) > 0)
    has_ftp_test_overlap = int(len(ftp_test_hits) > 0)
    has_ptp_test_overlap = int(len(ptp_test_hits) > 0)

    ftp_class_overlap_count = len(ftp_class_hits)
    ptp_class_overlap_count = len(ptp_class_hits)
    ftp_test_overlap_count = len(ftp_test_hits)
    ptp_test_overlap_count = len(ptp_test_hits)

    ftp_class_overlap_ratio = overlap_ratio(ftp_class_hits, ftp_sets["classes"])
    ptp_class_overlap_ratio = overlap_ratio(ptp_class_hits, ptp_sets["classes"])
    ftp_test_overlap_ratio = overlap_ratio(ftp_test_hits, ftp_sets["tests"])
    ptp_test_overlap_ratio = overlap_ratio(ptp_test_hits, ptp_sets["tests"])

    # -------------------------
    # Level 3: 时序级
    # -------------------------
    ftp_first_file_hit_step = first_hit_step(
        step_records,
        lambda rec: len(rec["text_paths"] | rec["tool_paths"] & ftp_sets["files"]) > 0
    )

    ptp_first_file_hit_step = first_hit_step(
        step_records,
        lambda rec: len(rec["text_paths"] | rec["tool_paths"] & ptp_sets["files"]) > 0
    )

    ftp_first_class_hit_step = first_hit_step(
        step_records,
        lambda rec: any(c.lower() in rec["text"] for c in ftp_sets["classes"])
    )

    ptp_first_class_hit_step = first_hit_step(
        step_records,
        lambda rec: any(c.lower() in rec["text"] for c in ptp_sets["classes"])
    )

    ftp_first_test_hit_step = first_hit_step(
        step_records,
        lambda rec: any(t.lower() in rec["text"] for t in ftp_sets["tests"])
    )

    ptp_first_test_hit_step = first_hit_step(
        step_records,
        lambda rec: any(t.lower() in rec["text"] for t in ptp_sets["tests"])
    )

    num_steps = max(len(step_records), 1)

    ftp_first_file_hit_step_norm = ftp_first_file_hit_step / num_steps if ftp_first_file_hit_step >= 0 else np.nan
    ptp_first_file_hit_step_norm = ptp_first_file_hit_step / num_steps if ptp_first_file_hit_step >= 0 else np.nan
    ftp_first_class_hit_step_norm = ftp_first_class_hit_step / num_steps if ftp_first_class_hit_step >= 0 else np.nan
    ptp_first_class_hit_step_norm = ptp_first_class_hit_step / num_steps if ptp_first_class_hit_step >= 0 else np.nan
    ftp_first_test_hit_step_norm = ftp_first_test_hit_step / num_steps if ftp_first_test_hit_step >= 0 else np.nan
    ptp_first_test_hit_step_norm = ptp_first_test_hit_step / num_steps if ptp_first_test_hit_step >= 0 else np.nan

    ftp_hit_before_submit_proxy = int(ftp_first_file_hit_step >= 0)
    ptp_hit_before_submit_proxy = int(ptp_first_file_hit_step >= 0)

    return pd.Series({
        # level 1
        "has_ftp_file_overlap": has_ftp_file_overlap,
        "has_ptp_file_overlap": has_ptp_file_overlap,
        "ftp_file_overlap_count": ftp_file_overlap_count,
        "ptp_file_overlap_count": ptp_file_overlap_count,
        "ftp_file_overlap_ratio": ftp_file_overlap_ratio,
        "ptp_file_overlap_ratio": ptp_file_overlap_ratio,

        # level 2
        "has_ftp_class_overlap": has_ftp_class_overlap,
        "has_ptp_class_overlap": has_ptp_class_overlap,
        "has_ftp_test_overlap": has_ftp_test_overlap,
        "has_ptp_test_overlap": has_ptp_test_overlap,
        "ftp_class_overlap_count": ftp_class_overlap_count,
        "ptp_class_overlap_count": ptp_class_overlap_count,
        "ftp_test_overlap_count": ftp_test_overlap_count,
        "ptp_test_overlap_count": ptp_test_overlap_count,
        "ftp_class_overlap_ratio": ftp_class_overlap_ratio,
        "ptp_class_overlap_ratio": ptp_class_overlap_ratio,
        "ftp_test_overlap_ratio": ftp_test_overlap_ratio,
        "ptp_test_overlap_ratio": ptp_test_overlap_ratio,

        # level 3
        "ftp_first_file_hit_step": ftp_first_file_hit_step,
        "ptp_first_file_hit_step": ptp_first_file_hit_step,
        "ftp_first_class_hit_step": ftp_first_class_hit_step,
        "ptp_first_class_hit_step": ptp_first_class_hit_step,
        "ftp_first_test_hit_step": ftp_first_test_hit_step,
        "ptp_first_test_hit_step": ptp_first_test_hit_step,
        "ftp_first_file_hit_step_norm": ftp_first_file_hit_step_norm,
        "ptp_first_file_hit_step_norm": ptp_first_file_hit_step_norm,
        "ftp_first_class_hit_step_norm": ftp_first_class_hit_step_norm,
        "ptp_first_class_hit_step_norm": ptp_first_class_hit_step_norm,
        "ftp_first_test_hit_step_norm": ftp_first_test_hit_step_norm,
        "ptp_first_test_hit_step_norm": ptp_first_test_hit_step_norm,
        "ftp_hit_before_submit_proxy": ftp_hit_before_submit_proxy,
        "ptp_hit_before_submit_proxy": ptp_hit_before_submit_proxy,
    })
```

```python
alignment_features_df = analysis_df.apply(make_alignment_features, axis=1)
alignment_df = pd.concat(
    [
        analysis_df[["instance_id", "traj_id", "repo_id", "repo", "resolved"]].reset_index(drop=True),
        alignment_features_df.reset_index(drop=True),
    ],
    axis=1
)

print(alignment_df.shape)
alignment_df.head()
```

```python
traj_out = RESULTS_DIR / "tool_alignment_features_traj_level.csv"
alignment_df.to_csv(traj_out, index=False)

print("Saved:", traj_out)
```

```python
instance_alignment_df = (
    alignment_df.groupby("instance_id", as_index=False)
    .agg(
        repo_id=("repo_id", "first"),
        repo=("repo", "first"),
        n_trajs=("traj_id", "count"),
        success_rate=("resolved", "mean"),
        any_success=("resolved", "max"),

        has_ftp_file_overlap=("has_ftp_file_overlap", "mean"),
        has_ptp_file_overlap=("has_ptp_file_overlap", "mean"),
        ftp_file_overlap_count=("ftp_file_overlap_count", "mean"),
        ptp_file_overlap_count=("ptp_file_overlap_count", "mean"),
        ftp_file_overlap_ratio=("ftp_file_overlap_ratio", "mean"),
        ptp_file_overlap_ratio=("ptp_file_overlap_ratio", "mean"),

        has_ftp_class_overlap=("has_ftp_class_overlap", "mean"),
        has_ptp_class_overlap=("has_ptp_class_overlap", "mean"),
        has_ftp_test_overlap=("has_ftp_test_overlap", "mean"),
        has_ptp_test_overlap=("has_ptp_test_overlap", "mean"),
        ftp_class_overlap_count=("ftp_class_overlap_count", "mean"),
        ptp_class_overlap_count=("ptp_class_overlap_count", "mean"),
        ftp_test_overlap_count=("ftp_test_overlap_count", "mean"),
        ptp_test_overlap_count=("ptp_test_overlap_count", "mean"),
        ftp_class_overlap_ratio=("ftp_class_overlap_ratio", "mean"),
        ptp_class_overlap_ratio=("ptp_class_overlap_ratio", "mean"),
        ftp_test_overlap_ratio=("ftp_test_overlap_ratio", "mean"),
        ptp_test_overlap_ratio=("ptp_test_overlap_ratio", "mean"),

        ftp_first_file_hit_step_norm=("ftp_first_file_hit_step_norm", "mean"),
        ptp_first_file_hit_step_norm=("ptp_first_file_hit_step_norm", "mean"),
        ftp_first_class_hit_step_norm=("ftp_first_class_hit_step_norm", "mean"),
        ptp_first_class_hit_step_norm=("ptp_first_class_hit_step_norm", "mean"),
        ftp_first_test_hit_step_norm=("ftp_first_test_hit_step_norm", "mean"),
        ptp_first_test_hit_step_norm=("ptp_first_test_hit_step_norm", "mean"),
    )
)

instance_out = RESULTS_DIR / "tool_alignment_features_instance_level.csv"
instance_alignment_df.to_csv(instance_out, index=False)

print("Saved:", instance_out)
print(instance_alignment_df.shape)
```
